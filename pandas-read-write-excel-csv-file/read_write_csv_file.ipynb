<h1>Notebook to learn How to perform read and write operations on .csv and .xlsx files with Pandas</h1>

<h3><b>Reading the CSV File</b></h3>

```python
import pandas as pd
```

```python
# let us get our data frame ready
df = pd.read_csv("netflix_titles.csv")
df
```

```python
df = pd.read_csv("netflix_titles.csv",header=None, nrows=3)
df
# header = None gives us our output without header names
```

```python
df = pd.read_csv("netflix_titles.csv", na_values=["not available","na","NA"])
df
# na_values = [] this argument will give us 'NaN' in output wherever the data is 
# either not available or missing as specified in the list items....
```

```python
df = pd.read_csv("netflix_titles.csv", na_values={
    'director':["not available","NA"],
    'cast':["not available","NA"]
})
df
# we can also na_values = {} as dictionary so that to get desired values in place of 
# missing data or imperfect data as shown in example
''' 
this technique comes in handy when we have different types of columns having different types of data
and we need to identify and sort that data based on our terms and conditions.
'''
```

<h3><b>Writing to a csv file</b></h3>

```python
df.to_csv("new_netflix.csv",index=False)
# this is the line to create a new file named "new_netflix.csv" and 
# the data of the "netflix_titles.csv" gets written into newly created "new_netflix.csv".

# we can check out the directory to see if it's made or not?. 

# when no argument other than filename is provided then it also adds index numbers to our rows containing data
# and index number starts from 0. 
# To get rid of the index numbers, we can write index=False, as second argument to this function.
# similarly, if we don't need headers in our new file, we can simply write as
#  header=False after/before index argument....

# by removing indexes from the file we can see that the new file is exact copy of our earlier file and 
# also its size is now same as that of earlier.
```

```python
# to take a look at our columns present in our csv file we can use this command
# df.columns
# using this command we can select the columns we want to write to our new csv file

# suppose we need to write only selected columns to our csv file,
# we can do that so as shown
df.to_csv("new_titles.csv",index=False,columns=['show_id','title','country','duration','listed_in'])
```
